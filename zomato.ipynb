convert rating 4/5 to 4

```python
def handlerate(rate):
  rate = str(rate).split('/')
  rate = rate[0];
  return float(rate)

df['rate'] = df['rate'].apply(handlerate)

print(df.head())
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('type of restaurant')
```
