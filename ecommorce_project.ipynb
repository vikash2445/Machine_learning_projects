```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin-1')
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

change date_time formet

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data.info()
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order day the week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
fig = px.line(sales_by_month, x='Order Month', y= 'Sales', title = "monthly sales analysis")

fig.show()
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
fig = px.pie(sales_by_category, values="Sales", names="Category",hole=0.5, color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()

```

Sales analysis by sub category

```python
Sales_by_sub_category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
Sales_by_sub_category
```

```python
fig = px.bar(Sales_by_sub_category, x='Sub-Category', y='Sales', title='Sales by Sub-Category')
fig.show()
```

Monthly profit analysis

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
fig = px.line(profit_by_month, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

Profit by category

```python
Profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
Profit_by_category
```

```python
fig = px.bar(Profit_by_category, x='Category', y='Profit', title='Profit by Category')
fig.show()
```

```python
profit_by_sub_category = data.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_sub_category
```

```python
fig = px.bar(profit_by_sub_category, x='Sub-Category', y='Profit', title='Profit by Sub-Category')
fig.show()
```
